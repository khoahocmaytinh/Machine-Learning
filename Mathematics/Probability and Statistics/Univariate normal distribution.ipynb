```
import numpy as np
import matplotlib.pyplot as plt
```

```
def f(x, mean, sigma):
  return np.exp(-(x - mean)**2/(2*sigma**2))/(np.sqrt(2*np.pi*sigma**2))
```

```
x = np.linspace(-10, 10, 1000)
px1 = f(x, -1, 2)
px2 = f(x, 0, 1)
px3 = f(x, .5, .5)

plt.plot(x, px1)
plt.plot(x, px2)
plt.plot(x, px3)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.show()
```
